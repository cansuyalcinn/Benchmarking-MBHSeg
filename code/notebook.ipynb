# add the nnunet nameing vs original names for the train val and test set for the convenience

```python
import os
import json

mbhseg24_splits = '/home/cansu/Benchmarking-MBHSeg/data/splits/mbhseg24'
mbhseg25_splits = '/home/cansu/Benchmarking-MBHSeg/data/splits/mbhseg25'

fold = 4
dataset_name = 'Mbhseg24'

# read train and val splits
with open(os.path.join(mbhseg24_splits, f'fold_{fold}_train.txt'), 'r') as f:
    mbhseg24_train = f.read().splitlines()

with open(os.path.join(mbhseg24_splits, f'fold_{fold}_val.txt'), 'r') as f:
    mbhseg24_val = f.read().splitlines()

# nnunet names vs original names dictionary for all data
trainval_names = '/home/cansu/Benchmarking-MBHSeg/code/data_nnunet/nnUNet_raw/Dataset200_Mbhseg24/dict_names_trainval.json'

# read the existing dictionary
with open(trainval_names, 'r') as f:
    dict_names_trainval = json.load(f)

# Reverse the dictionary to map original ID to nnunet name for matching
dict_names_reversed = {v: k for k, v in dict_names_trainval.items()}

# create nested dictionary for train and val with nnunet names as keys (without _0000) and original names as values
nested_dict = {
    "train": {dict_names_reversed[k].rsplit('_', 1)[0]: dict_names_trainval[dict_names_reversed[k]] for k in mbhseg24_train if k in dict_names_reversed},
    "val": {dict_names_reversed[k].rsplit('_', 1)[0]: dict_names_trainval[dict_names_reversed[k]] for k in mbhseg24_val if k in dict_names_reversed}
}

# save to a new json file
output_file = f'/home/cansu/Benchmarking-MBHSeg/code/data_nnunet/all_splits_trainset/{dataset_name}_dict_names_trainval_fold_{fold}.json'
with open(output_file, 'w') as f:
    json.dump(nested_dict, f, indent=4)

print(f"Nested dictionary saved to {output_file}")
```
